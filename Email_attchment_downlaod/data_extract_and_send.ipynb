```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
data=pd.read_excel("/home/muhammadshoaib/Downloads/HS_Agreed_Template (1).xlsx")
```

```python
Inventory_Item = []
Product_Code = []
IMEI = []
From_Location = []
To_Location = []
Quantity = []
Department = []
Request_type = []
PO_Number = []
Receipt_No = []
Action = []
Sub_Inventory = []

new_data = {
    "Inventory_Item": Inventory_Item,
    "Product_Code": Product_Code,
    "IMEI": IMEI,
    "From_Location": From_Location,
    "To_Location": To_Location,
    "Quantity": Quantity,
    "Department": Department,
    "Request_type": Request_type,
    "PO_Number": PO_Number,
    "Receipt_No": Receipt_No,
    "Action": Action,
    "Sub_Inventory": Sub_Inventory
}
```

```python
def extract_data(data):
    # counter = 0
    for index, d in data.iterrows():
        if d['IMEI'] in IMEI:
            # If duplicate IMEI is found, drop the entire row
            data.drop(index, inplace=True)
        else:
        # if counter <= 10:
            for column_name, value in d.items():
                if column_name == "Inventory_Item":
                    Inventory_Item.append(value)
                elif column_name == "Product_Code":
                    Product_Code.append(value)
                elif column_name == "IMEI":
                    IMEI.append(value)
                elif column_name == "From_Location":
                    From_Location.append(value)
                elif column_name == "To_Location":
                    To_Location.append(value)
                elif column_name == "Quantity":
                    Quantity.append(value)
                elif column_name == "Department":
                    Department.append(value)
                elif column_name == "Request type":
                    Request_type.append(value)
                elif column_name == "PO_Number":
                    PO_Number.append(value)
                elif column_name == "Receipt_No":
                    Receipt_No.append(value)
                elif column_name == "Action":
                    Action.append(value)
                elif column_name == "Sub Inventory":
                    Sub_Inventory.append(value)
                else:
                    # If the column name doesn't match any expected values, print a message
                    print("Encountered unexpected column name:", column_name)
            # counter += 1

extract_data(data)

```

```python
new_data_frame = pd.DataFrame(new_data)
```
